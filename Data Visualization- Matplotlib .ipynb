# Data Visualization 

Data visualization is the graphical representation of data and information using visual elements such as charts, graphs, maps, and other visual aids. It is a powerful tool for making data more accessible, understandable, and interpretable. Data visualization allows you to convey complex information in a visually appealing and easily comprehensible format, making it a key component in data analysis, decision-making, and communication.

 # Here are some common libraries and tools used for data visualization:


### Matplotlib:

Matplotlib is one of the most popular and widely used data visualization libraries in Python.
It provides a wide range of 2D and 3D plotting capabilities, allowing you to create a variety of static, animated, or interactive visualizations.

### Seaborn:

Seaborn is a Python data visualization library built on top of Matplotlib.
It is particularly useful for creating statistical graphics and provides functions for easily visualizing complex datasets.

### Plotly:

Plotly is an interactive data visualization library that allows you to create web-based, interactive visualizations and dashboards.
It supports a wide range of chart types and is often used for creating dynamic and interactive plots.

### Bokeh:

Bokeh is another Python library for creating interactive and web-based visualizations.
It is well-suited for creating interactive, data-driven applications and dashboards.

```python
x = np.linspace(0,10,200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("ths is my sin graph")
plt.show()
```

```python
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y, c = colours , s= sizes , alpha= .5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Random Graph')
plt.show()
```

```python
x = ['a', 'b' , 'c', 'd' ,'e']
```

```python
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(" This is my Bar Chart")
```

```python
y = np.random.rand(5)
plt.barh(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(" This is my Bar Chart")
```

```python
plt.figure(figsize= (6,2))
plt.plot(x,y , '--r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(' sin graph')
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c= 'red' , s=sizes , alpha =.8 , marker='v')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('V Graph')
plt.show()
```
